```python
import qutip as qt
import pulsee.plot as psplt
import operators as op
from qutip import Qobj
import numpy as np
```

```python
SFO1 = 202.4744410 * 10 ** 6
SFO2 = 202.4751948 * 10 ** 6
SF = 202.4744410 * 10 ** 6
omega_1 = SFO1 - SF
omega_2 = SFO2 - SF
J = 22
t = 1 / (2 * J)
omega_2
```

```python
#ignoring all zeeman terms (this is also the B=0 case)

H_paper = np.pi * J * 2 * op.IzSz  #from our paper
U_J_paper = qt.Qobj(-1j * H_paper * t).expm()

U_J_levitt = qt.Qobj((-1j * np.pi * op.IzSz)).expm()  # J coupling with wait time t = 1/(2J). Levitt pg 398

U_J_oliveira = op.U_J(t, J)  #Oliveira with zeeman terms ignored, pg 66

H_strong = np.pi * J * 2 * op.IzSz + 2 * np.pi * J * (op.IxSx + op.IySy)  # J coupling strong coupling limit ime t = 1/(2J). Starting Levitt pg 615
U_J_strong = qt.Qobj(-1j * H_strong * t).expm()

print("U_J_paper")
display(U_J_paper)
psplt.plot_complex_density_matrix(U_J_paper)

print("U_J_levitt")
display(U_J_levitt)
psplt.plot_complex_density_matrix(U_J_levitt)

print("U_J_oliveira")
display(U_J_oliveira)
psplt.plot_complex_density_matrix(U_J_oliveira)

print("U_J_strong")
display(U_J_strong)
psplt.plot_complex_density_matrix(U_J_strong)
```

```python
#including back all zeeman terms
H_paper = omega_1 * op.Iz + omega_2 * op.Sz + np.pi * J * 2 * op.IzSz  #from our paper
U_J_paper = qt.Qobj(-1j * H_paper * t).expm()

U_J_levitt = qt.Qobj((-1j * np.pi * op.IzSz)).expm()  # J coupling with wait time t = 1/(2J). Levitt pg 398
U_J_levitt_chemical_shift_1 = qt.Qobj((-1j * omega_1 * t * op.Iz)).expm()
U_J_levitt_chemical_shift_2 = qt.Qobj((-1j * omega_2 * t * op.Sz)).expm()
U_J_levitt_full = U_J_levitt_chemical_shift_1 * U_J_levitt_chemical_shift_2 * U_J_levitt

U_J_oliveira = op.U_J(t, J)  #Oliveira with zeeman terms ignored, pg 66
U_J_oliveira_full = U_J_levitt_chemical_shift_1 * U_J_levitt_chemical_shift_2 * U_J_oliveira  #including Levitt's convention because Oliveira double resonance frame is unclear

H_strong = omega_1 * op.Iz + omega_2 * op.Sz + np.pi * J * 2 * op.IzSz + 2 * np.pi * J * (
            op.IxSx + op.IySy)  # J coupling strong coupling limit ime t = 1/(2J). Starting Levitt pg 615
U_J_strong = qt.Qobj(-1j * H_strong * t).expm()

print("U_J_paper")
display(U_J_paper)
psplt.plot_complex_density_matrix(U_J_paper)

print("U_J_levitt_full")
display(U_J_levitt_full)
psplt.plot_complex_density_matrix(U_J_levitt_full)

print("U_J_oliveira_full")
display(U_J_oliveira_full)
psplt.plot_complex_density_matrix(U_J_oliveira_full)

print("U_J_strong")
display(U_J_strong)
psplt.plot_complex_density_matrix(U_J_strong)
```

```python
cnot_pseudo = np.sqrt(1j) * op.Rx_S(np.pi / 2) * U_J_levitt * op.Ry_S(np.pi / 2)

print("cnot_pseudo")
display(cnot_pseudo)
psplt.plot_complex_density_matrix(cnot_pseudo)

cnot_pseudo_full = np.sqrt(1j) * op.Rx_S(np.pi / 2) * U_J_levitt_full * op.Ry_S(np.pi / 2)

print("cnot_pseudo_full")
display(cnot_pseudo_full)
psplt.plot_complex_density_matrix(cnot_pseudo_full)
```

```python
#calculate for Earth's field
# since 0 = B^0 Levitt pg. 29

omega_1 = 0
omega_2 = (SFO2 - SF) * (5e-5 / 11.7467)
#omega_2=0 #switch to Earths field=0

H_strong = op.Sz + np.pi * J * 2 * op.IzSz + 2 * np.pi * J * (
            op.IxSx + op.IySy)  # J coupling strong coupling limit ime t = 1/(2J). Starting Levitt pg 615
U_J_strong = qt.Qobj(-1j * H_strong * t).expm()

H_strong_full = omega_1 * op.Iz + omega_2 * op.Sz + np.pi * J * 2 * op.IzSz + 2 * np.pi * J * (
            op.IxSx + op.IySy)  # J coupling strong coupling limit ime t = 1/(2J). Starting Levitt pg 615
U_J_strong_full = qt.Qobj(-1j * H_strong_full * t).expm()

print("U_J_strong")
display(U_J_strong)
psplt.plot_complex_density_matrix(U_J_strong)

print("U_J_strong_full")
display(U_J_strong_full)
psplt.plot_complex_density_matrix(U_J_strong_full)
```

```python
cnot_pseudo_strong = np.sqrt(1j) * op.Rx_S(np.pi / 2) * U_J_strong * op.Ry_S(np.pi / 2)
#cnot_pseudo_strong is the CNOT gate in strong coupling
print("cnot_pseudo_strong ")
display(cnot_pseudo_strong)
psplt.plot_complex_density_matrix(cnot_pseudo_strong)

cnot_pseudo_full_strong = np.sqrt(1j) * op.Rx_S(np.pi / 2) * U_J_strong_full * op.Ry_S(np.pi / 2)

print("cnot_pseudo_full_strong")
display(cnot_pseudo_full_strong)
psplt.plot_complex_density_matrix(cnot_pseudo_strong)
```

```python
cnot_correct = op.Rz_I(np.pi/2) * op.Rz_S(-np.pi/2) * cnot_pseudo
psplt.plot_complex_density_matrix(cnot_correct)
#this is the gate we want to recreate with the strong coupling Hamiltonian
```

```python
#lets figure out what matrix makes this transformation and solve X=TY matrix equation

# Example matrices
Y = U_J_strong.full()  # This is the matrix we have
X = cnot_correct.full()   # This is the matrix we want to create

# T is the transformation matrix

# Check if L is invertible
if np.linalg.det(Y) == 0:
    print("Matrix Y is not invertible.")
else:
    # Compute the inverse of Y
    Y_inv = np.linalg.inv(Y)
    
    # Solve for T
    T = np.dot(Y_inv, X)
    print("Solution T:")
    print(T)
```

```python
T_qt_object=Qobj(T,dims=[[2, 2], [2, 2]])
psplt.plot_complex_density_matrix(T_qt_object)
print(T_qt_object.dag()*T_qt_object)
```
