```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.cm as cm
# Define the Viridis colormap with 15 equal steps
viridis = cm.get_cmap('viridis', 15)

# Extract the colors from the Viridis colormap
viridis_colors = [mcolors.to_hex(viridis(i)) for i in range(15)]

# Plot the Viridis color palette in a horizontal bar with hex codes
fig, ax = plt.subplots(figsize=(15, 2))

# Create a color bar to display each hex code as a color patch with text
for i, color in enumerate(viridis_colors):
    ax.add_patch(plt.Rectangle((i, 0), 1, 1, color=color))
    ax.text(i + 0.5, -0.3, f"{color}", ha='center', va='center', fontsize=10, color='black')

# Remove ticks and labels for clarity
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlim(0, len(viridis_colors))
ax.set_aspect("equal")

# Show the plot
plt.show()
```
